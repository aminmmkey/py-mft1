```python
import time
import urllib.request

link = "https://api.pro.coinbase.com/products/BTC-USD/candles/1h"
req = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/66.0.2'})

req = urllib.request.urlopen(req)
html = req.read()
text = html.decode('utf-8')
data = eval(text)
data
```

```python
arr = np.array(data)
```

```python
t = arr[:, 0]
vol = arr[:, -1]
indx = range(300)
```

```python
plt.bar(indx[-24:], vol[-24:])
```

```python
for i in [1, 2, 3, 4]:
    plt.plot(indx[-24:],arr[:, i][-24:])
```

```python
plt.scatter(vol, arr[:, 3], c=t)
plt.colorbar()
```
